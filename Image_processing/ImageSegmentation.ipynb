```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.data import camera, coins
from skimage.filters import threshold_multiotsu
```

```python
image = camera()
```

```python
plt.imshow(image,cmap='gray')
```

```python
plt.hist(image)
```

```python
image[(image>150) & (image<=255)] = 255
image[(image>70) & (image<=150)] = 150
image[(image>0) & (image<=70)] = 1
```

```python
plt.imshow(image,cmap='gray')
```
